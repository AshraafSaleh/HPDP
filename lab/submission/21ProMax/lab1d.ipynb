Please use the [**Chipotle Sales dataset**](https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv) for this exercise.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
url = 'https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv'
df = pd.read_csv(url, sep = '\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
```

### Step 4. How many products cost more than $10.00?

```python
chipo.head()
```

```python
chipo.info()
```

```python
pd.to_numeric?
```

```python
chipo['price_float'] = pd.to_numeric(chipo.item_price.str.slice(1))
```

```python
(chipo['price_float'] > 10).sum()
```

```python
product_prices = chipo.groupby('item_name').agg({'price_float': max})
```

```python
product_prices.info()
```

```python
(product_prices.price_float > 10).sum()
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo['price_per_item'] = chipo.price_float / chipo.quantity
```

```python
product_prices = chipo.groupby('item_name').agg({'price_per_item': max})
```

```python
(product_prices.price_per_item > 10).sum()
```

```python
product_prices
```

### Step 6. Sort by the name of the item

```python
product_prices.sort_index()
```
